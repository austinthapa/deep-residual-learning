```python
import tensorflow as tf

from tensorflow.keras import layers
from functools import partial
```

```python
tf.__version__
```

```python
conv_2d = partial(
    tf.keras.layers.Conv2D, 
    kernel_size = (3, 3),
    strides = 1, 
    padding = "same",
    activation = "relu",
    kernel_initializer = "he_normal"
)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(layers.Input(shape = (224, 224, 3)))
model.add(conv_2d(filters = 64, kernel_size = (7, 7), strides = 2))
model.add(layers.BatchNormalization())
model.add(layers.ReLU())
```

```python
for _ in range(6):
    model.add(conv_2d(filters = 64))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
```

```python
model.add(conv_2d(filters = 128, strides = 2))
model.add(layers.BatchNormalization())
model.add(layers.ReLU())
```

```python
for _ in range(7):
    model.add(conv_2d(filters = 128))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
```

```python
model.add(conv_2d(filters = 256, strides = 2))
model.add(layers.BatchNormalization())
model.add(layers.ReLU())
```

```python
for _ in range(11):
    model.add(conv_2d(filters = 256))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
```

```python
model.add(conv_2d(filters = 512, strides = 2))
model.add(layers.BatchNormalization())
model.add(layers.ReLU())
```

```python
for _ in range(5):
    model.add(conv_2d(filters = 512))
    model.add(layers.BatchNormalization())
    model.add(layers.ReLU())
```

```python
model.add(layers.GlobalAveragePooling2D())
model.add(layers.Flatten())
model.add(layers.Dense(1000, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(
    loss = "sparse_categorical_crossentropy",
    optimizer = "adam",
    metrics = ["accuracy"]
)
```
